```python
#list manipulation game

#instantiate
current = ["apple", "banana", "cookie", "xylophone", "zebra"]

def make_target_list(current_list):
    target = [current_list[4], current_list[0]] 
    return target

def display_lists(current_list, target_list):
    print("Current list:", current_list)
    print("Target list:", target_list)

def operations(current_list):
    user_operation = input("Would you like to append, extend, concatenate, traverse, modify, insert, pop, sort, or remove from the list? \nWrite 'append', 'extend', 'concatenate', 'traverse', 'modify', 'insert', 'pop', 'sort', or 'remove'.")

#append
    if user_operation.lower() == "append":
        word_append = input("What word would you like to append?")
        current_list = current_list.append(word_append)
#extend
    elif user_operation.lower() == "extend":
        list_extend = input("Write out a list separated by commas that you want to extend to your current list.")
        current_list = current_list.extend(list_extend)
#concatenate
    elif user_operation.lower() == "concatenate":
        words_concat_str = input("Write out the two words you want to concatenate, comma-separated.")
        words_concat = words_concat_str.split(",")
        for item in current_list:
            if item in words_concat:
                current_list.remove(item)
            else:
                continue
#traverse
    elif user_operation.lower() == "traverse":
        for item in current_list:
            print(item)
#modify
    elif user_operation.lower() == "modify":
        index_modify = int(input("What is the index of the word you want to modify? Give the index, where 0 would mean 1st word, 1 would mean second word, and so on."))
        word_modify_to = input("What is the word you want to change it to?")
        current_list[index_modify] = word_modify_to
        print(current_list)

#insert
    elif user_operation.lower() == "insert":
        word_insert = input("What word do you want to insert into the list?")
        where_insert = int(input("Where in the list do you want to insert it? Give the index, where 0 would mean 1st word, 1 would mean second word, and so on."))
        current_list = current_list.insert(where_insert, word_insert)
        print("Current list:", current_list)
#pop
    elif user_operation.lower() == "pop":
        user_pop = input("What word do you want to pop? (I.e. remove from list and name as separate variable)")
        word_pop = current_list.pop(user_pop)
        print("Current list:", current_list)
        print("Popped word:", word_pop)
#sort
    elif user_operation.lower() == "sort":
        while user_operation == "sort":
            list_sort = input("Ascending or descending order? Enter A or D.")
            if list_sort == "A" or list_sort == "a":
                current_list = current_list.sort()
                break
            elif list_sort == "D" or list_sort == "d":
                current_list = current_list.sort(reverse=True)
                break
            else:
                print("Try again.")
#remove
    elif user_operation.lower() == "remove":
        list_remove = input("What word would you like to remove from the list?")
        if list_remove in current_list:
            current_list = current_list.remove(list_remove)
        else:
            print("That word isn't in the list.")

    
    else:
        print("Not a valid option.")
    
    return


def compare_lists(current_list, target_list):
    if current_list == target_list:
        True
        print("You got it! Congrats!")
    else:
        False
        print("Not quite there!")
```

```python
t = []

t = make_target_list(current)
display_lists(current, t)
while current != t:
    operations(current)
    display_lists(current, t)
# compare_lists(current, t)
```
